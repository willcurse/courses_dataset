```python
numeric_col=['Age','GPA']
fig,axes=plt.subplots(1,2,figsize=(12,5))
for index,column in enumerate(numeric_col):
    sns.kdeplot(data2[column],ax=axes[index])
    axes[index].set_title(f'KDE plot of {column}')
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import shapiro

shapiro_result = {}

for column in numeric_col:
    stat, p_value = shapiro(data2[column])
    shapiro_result[column] = round(p_value, 3)

shapiro_result
```

```python

def log_trans(data, col_name):
    data[f'{col_name}_log'] = np.log(data[col_name])
    stat, p_value = shapiro(data[f'{col_name}_log'])
    kdeplot = sns.kdeplot(data[f'{col_name}_log'])
    print(kdeplot)
    print(f"Shapiro-Wilk test p-value: {p_value}")
```

```python
log_trans(data2,'Age')
```

```python
from scipy.stats import boxcox

def boxcox_trans(data,col_name):
    transfromed_data,_=boxcox(data[col_name])
    data[f'{col_name}']=transfromed_data
    stat,p_value=shapiro(data[f'{col_name}'])
    kdeplot=sns.kdeplot(data[f'{col_name}'])
    
    print(kdeplot)
    print(p_value)
```

```python
boxcox_trans(data2,'GPA')
```

```python
from scipy.stats import yeojohnson

def yeojohnson_trans(data,col_name):
    transfromed_data,_=yeojohnson(data[col_name])
    data[f'{col_name}']=transfromed_data
    stat,p_value=shapiro(data[f'{col_name}'])
    kdeplot=sns.kdeplot(data[f'{col_name}'])
    
    print(kdeplot)
    print(p_value)
```

```python
yeojohnson_trans(data2,'GPA')
```

```python
from scipy import stats
hypothesis_mean= 3.0
t_stat,p_value=stats.ttest_1samp(data2['GPA'],hypothesis_mean)
print("p_value: ",p_value)

if p_value<0.05:
    print("reject the null hypothesis")
else:
    print("failed")
```
